```python
# Part 2 Redo
# The mathy, range-only way so my computer doesn't die on me
# Reference: https://topaz.github.io/paste/#XQAAAQCbCgAAAAAAAAARiEJHiiMzw3cPM/1Vl+2nx/DqKkM2yi+AsNSFTsnyEq6tvQ1/bKc1NBj3YF9MoLQJaXX5mzxYAIeb8s6xFC494l5BU4SjxJ7raCQWlBCg3cHqOkdhOuJLhT2Nj3yE8CvXqI6WnMQN5lC9Q1iIW1Ivcv76M49JdzKfryCcJJ4fDhAPOkSflcV9wjxoOuKiWzQd3Vzf1+pQn3UCEQYswytRrpAAZ0IiloE732L1L4u1Z2IbDs6MStJqzd1K04kwfmEcb8SEGD983LfhOpfFHluyGPhJU+5ympTfnhe7FMrz1DqW/OSCiNOKI3rv6AXZs7ShMY6pAbmY/EEyeyPQsI190dSapcpmKxzR83if5WixrQMpMiZ5bW/x240MKY6UMhctaDyXt90f5wihfoNUpVUEpVO9MZAB1N5dZG07WrI5yIWBK5QKRn47gTx3Y2oR9RBKCYc7uScNtUWfewqxlSKwHBlCJsg1EU2kYTWhCSe/0FCSDimGa6V5/dRXuimP9NV+AIBMxqlGgDELbXBJtuCr0awmFOmqWAVXBsAZuyLOWVcbELdckJLQRqs+T+1qJneZiRife58WGZ+Sfoyx3u2NzMxxKDDrz+tRZ5GfNV+BVJRhEts+K0EIOPqqLLvdngxp78lybxWCjK/Xv+rX2GWluik+KDe7wban4BaiAZEeU7MyJL37VvqBXvUnEPUPMwgnD0KlCYfc/SdKPKwev4K6tc3GuojHPu+CBJ6ggz1MScNBbQIoDvPlQaVK03GIiHIni9omJgDA5XoETtxKJLd6wfy9bdos3+W61+KJ6QZpYUYJ7OWqIy7MXpnAcTlsxkDNArfLolLr4GQWVyzw+YOf5bxICq0XvK5uHxUJ5X+tDRrE4KhLxQRPMd/BDDsr4UEPqcIp4ojddlStf71ih7WEmb8UwDizepoAjej0s0Cm1lYMQJSl3gGo+LXKKNaRHceLHeWsb4E9ZJCcLbvNCL+XE+kp9prRfXq0byj+XlHqOY6I/BCiDpMuq8h5YyiXzMf1/+ui6cM=
arrayInput = stringInput.split("\n")
arrayTest = testString.split("\n")

curr_input = arrayInput

def input_parse():
    seeds = []
    maps = []

    for line in curr_input:
        line_type = line.split()

        # Get given map initializated
        if line_type[1] == 'map:':
            maps.append([])
            map_index = len(maps) - 1
            continue
        # Get given seed start and end values
        if line_type[0] == 'seeds:':
            i = 1
            while i < len(line_type):
                seed_start = int(line_type[i])
                seed_end = seed_start + int(line_type[i + 1]) - 1
                seeds.append((seed_start, seed_end))
                i += 2
            continue
        # Get given map start and end values
        dest = int(line_type[0])
        source = int(line_type[1])
        range_len = int(line_type[2])

        source_end = source + range_len - 1
        adjustment = dest - source # adjustment gets the difference between the dest and source
        maps[map_index].append((source, source_end, adjustment))
    return seeds, maps

def map_func(seeds, maps):
    lowest = 99999999999999999
    for seed in seeds: # for each seed range
        destinations = [] # destinations is a list of tuples of (start, end) of the seed range
        destinations.append(seed)
        for map in maps: # for each map
            sources = destinations # sources is a list of tuples of (start, end) of the seed range
            destinations = []

            # Go through each source and find the destination
            while len(sources) > 0: 
                source = sources.pop()
                source_start, source_end = source
                for range in map:    
                    map_start, map_end, adjust = range
                    # if the source is within the map range, add the destination to the destinations list
                    if source_start >= map_start and source_end <= map_end:
                        # append the source range with the adjustment to the destinations list
                        destinations.append((source_start+adjust, source_end+adjust))
                        break
                    # if the source is outside the map range, continue
                    if source_end < map_start or source_start > map_end:
                        continue
                    # if the source is partially within the map range, split the source into two and add to sources list
                    # We will later check if the spilt source is within any map range again
                    if source_start < map_start:
                        sources.append((source_start, map_start-1))
                        sources.append((map_start, source_end))
                        break
                    if source_end > map_end:
                        sources.append((source_start, map_end))
                        sources.append((map_end+1, source_end))
                        break
                else: # if the source is not within any map range, add it to the destinations list
                    destinations.append(source)
        # find the lowest location of the destinations
        seed_lowest = lowest_location(destinations)
        lowest = min(lowest, seed_lowest)
    return lowest

def lowest_location(locations):
    lowest = locations[0][0]
    for location in locations:
        lowest = min(location[0], lowest)
    return lowest

seeds, maps = input_parse()
print(seeds)
print(maps)
print(map_func(seeds, maps))
```
