```python
# Part 1: Determine the sum of the part numbers
# A part number is any number adjacent to a symbol, even diagonally

arrayInput = stringInput.split("\n")
arrayTest = testString.split("\n")

keyboard_symbols = [
    '!', '@', '#', '$', '%', '^', '&', '*', '(', ')',
    '-', '_', '=', '+', '[', '{', ']', '}', '\\', '|',
    ';', ':', "'", '"', ',', '<', '>', '/', '?',
    '`', '~', '`', '~'
]

number_index = []
total = 0 
symbol_index = []

# parse through and create dictory of numbers and their coordinates
for i in range(len(arrayInput)):
    engine_list = list(arrayInput[i])
    # print(engine_list)
    row_list = []
    last_item_num = False
    for j in range(len(engine_list)):
        if engine_list[j].isnumeric():
            if j != 0 and last_item_num:
                row_list[-1]["number"] += engine_list[j]
                row_list[-1]["symbol_coordinates"] += [[i-1, j-1], [i-1, j], [i-1, j+1], [i, j-1], [i, j+1], [i+1, j-1], [i+1, j], [i+1, j+1]]

            else:
                row_list.append({"number": engine_list[j], 
                                 "symbol_coordinates": [[i-1, j-1], [i-1, j], [i-1, j+1], [i, j-1], [i, j+1], [i+1, j-1], [i+1, j], [i+1, j+1]]})
            last_item_num = True
        else:
            last_item_num = False
            if engine_list[j] in keyboard_symbols:
                symbol_index.append([i, j])
    number_index.append(row_list)

# parase through number_index and determine if number is adjacent to symbol
for i in number_index:
    for j in i:
        for k in j["symbol_coordinates"]:
            if k in symbol_index:
                total += int(j["number"])
                break

print(total)
    
```

```python
# Part 2: Determine the sum of gear ratios
# A gear is any * symbol that is adjacent to exactly two part numbers. 
#Its gear ratio is the result of multiplying those two numbers together.


arrayInput = stringInput.split("\n")
arrayTest = testString.split("\n")

keyboard_symbols = [
    '!', '@', '#', '$', '%', '^', '&', '*', '(', ')',
    '-', '_', '=', '+', '[', '{', ']', '}', '\\', '|',
    ';', ':', "'", '"', ',', '<', '>', '/', '?',
    '`', '~', '`', '~'
]

number_index = []
total = 0 
symbol_index = []

# parse through and create dictory of numbers and their coordinates
for i in range(len(arrayInput)):
    engine_list = list(arrayInput[i])
    row_list = []
    star_list = []
    last_item_num = False
    for j in range(len(engine_list)):
        if engine_list[j].isnumeric():
            if j != 0 and last_item_num:
                row_list[-1]["number"] += engine_list[j]
                row_list[-1]["num_coordinates"] += [[i, j]]

            else:
                row_list.append({"number": engine_list[j], "num_coordinates" : [[i, j]]} )
            last_item_num = True
        else:
            last_item_num = False
            if engine_list[j] == "*":
                star_list.append({"part_placement" : [[i-1, j-1], [i-1, j], [i-1, j+1], [i, j-1], [i, j+1], [i+1, j-1], [i+1, j], [i+1, j+1]]})
    number_index.append(row_list)
    symbol_index.append(star_list)

# parase through symbol_index and determine if symbol is adjacent to number
for row in symbol_index:
    for symbol in row:
        num_counter = 0
        gear_ratio = 1
        numbers_used = []
        for k in symbol["part_placement"]:
            for num_row in number_index:
                for number in num_row:
                    if k in number["num_coordinates"] and number["number"] not in numbers_used:
                        num_counter += 1
                        gear_ratio *= int(number["number"])
                        numbers_used.append(number["number"])
        if num_counter == 2:
            total += gear_ratio

print(f'total = {total}')
    
```
